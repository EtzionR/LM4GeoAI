## Imports

```python
from transformers import pipeline
import pandas as pd
```

## Define NER model

For this tutorial, we would use [**distilbert-NER**](https://huggingface.co/dslim/distilbert-NER)

```python
MODEL = "dslim/bert-base-NER"
MODEL
```

## Activate NER model

```python
ner = pipeline("ner",
               model=MODEL,
               aggregation_strategy='average')

ner
```

## Example text for entity extraction

```python
example_text = "Sam Altman visited OpenAI in San Francisco."
example_text
```

## Apply NER model on the example text

```python
output = ner(example_text)
output
```

## Display the output as pandas Dataframe

```python
pd.DataFrame(output)
```
