## Import

```python
import folium
```

## Create a folium map

```python
main_y, main_x = 41.903725, 12.464165

fmap = folium.Map(location=[main_y, main_x], zoom_start=14.5, width=900, height=300)
fmap
```

## Add 3 Points to the map

Marker icons base on Glyphicons icons: [**getbootstrap**](https://getbootstrap.com/docs/3.3/components/#glyphicons)

```python
folium.Marker([41.902222, 12.457246], tooltip="St. Peter Obelisk").add_to(fmap)

folium.Marker([41.902912, 12.454416], tooltip="Sistine Chapel",
              icon=folium.Icon(color='green', icon='camera')).add_to(fmap)

folium.Marker([41.899809, 12.473027], tooltip="Neptune Fountain",
              icon=folium.Icon(color='red', icon='flag')).add_to(fmap)

fmap
```

## Add a single path from St. Peter to Neptune Fountain

```python
walk_path = [[41.902199, 12.457249], [41.902295, 12.459159], [41.902350, 12.462892],
             [41.900742, 12.464930], [41.901046, 12.465423], [41.901157, 12.465885],
             [41.901221, 12.466485], [41.900576, 12.466571], [41.900568, 12.467419],
             [41.900549, 12.468137], [41.900569, 12.468395], [41.900537, 12.469006],
             [41.900498, 12.470846], [41.900501, 12.471324], [41.900506, 12.471930],
             [41.900466, 12.473164], [41.899811, 12.473185], [41.899811, 12.473024]]

tooltip = 'Path from St. Peter to Neptune Fountain'

folium.PolyLine(locations=walk_path, color="purple", tooltip = tooltip).add_to(fmap)

fmap
```

## Add a polygon around Castel St. Angelo

```python
castel = [[41.905400, 12.4661949], [41.904465, 12.465304], [41.903962, 12.464043],
          [41.903236, 12.4646392], [41.902225, 12.465025], [41.902437, 12.465674],
          [41.902505, 12.4663880], [41.902617, 12.468131], [41.903284, 12.468093],
          [41.904286, 12.4686679], [41.904605, 12.467235], [41.905400, 12.466194]]

tooltip = 'Castel St. Angelo'

folium.Polygon(locations=castel, fill_color="blue", fill_opacity=0.25, tooltip=tooltip).add_to(fmap)

fmap
```
