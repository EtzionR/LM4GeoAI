```python
from geopy.geocoders import Nominatim
```

## Example for input location text

```python
example = 'London, United Kingdom'
example
```

## Init Nominatim object

```python
geolocator = Nominatim(user_agent="GeoAI_Course_Nominatim_Tutorial", timeout=10)
geolocator
```

## Geocode example location using Nominatim

```python
output = geolocator.geocode(example)
output
```

## Print Output result

```python
print(f'Location full adress:\n{output.address}\n')
print(f'Output coordiantes:\nWGS84 GEO X = {round(output.longitude,6)}, Y = {round(output.latitude,6)}')
```
